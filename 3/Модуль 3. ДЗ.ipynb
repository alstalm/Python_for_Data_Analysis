Для покупок из файла data_no_header.txt посчитаем метрику LTV (средняя сумма покупок на пользователя).

Учитываем всех пользователей независимо от количества их покупок.

LTV = сумма всех доходов / количество всех уникальных пользователей

Несколько замечаний по условию:

1. LTV можно посчитать как отношение суммы доходов к количеству пользователей. 

2. Под "суммой всех доходов на пользователя" понимается именно сумма, которую потратил пользователь на все свои покупки. 
3. Конечное отношение суммы доходов к количеству уникальных пользователей считаем посчитать после завершения цикла прохода по файлу. Вычисление средних значений LTV в процессе построчного чтения файла (и дальнейшее их усреднение) в рамках данной задачи приведет к неверному ответу.

```python
### LTV = сумма всех доходов / количество всех уникальных пользователей

def LTV_calculation( user_id_transactions ):
    amount_paid_for_user=0   
    id=0
    for user_id in user_id_transactions:
        amount_paid_for_user += user_id_transactions[id][1]
        id+=1
    return (amount_paid_for_user)

#user_id_transactions[число][1]. Здесь число это порядковый номер списка [dateTime, amount_paid] внутри user_id_transactions
```
